```python
import requests
from bs4 import BeautifulSoup

url = "https://www.crummy.com/software/BeautifulSoup/bs4/doc/"
html = requests.get(url)
soup = BeautifulSoup(html.content, 'lxml')

print(soup.prettify())
```
