Write a Python program to extract and display all the header tags from en.wikipedia.org/wiki/Main_Page.


```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
from time import sleep
from random import randint
import numpy as np

h1 = []

html = requests.get("https://en.wikipedia.org/wiki/Main_Page")
#print(html)
soup = BeautifulSoup(html.content, 'lxml')
header_tags = soup.find_all(["h1","h2","h3","h4","h5","h5"])
print("HEADER-TAGS",*header_tags , sep="\n\n")
sleep(randint(2,10))
```

```python
tag_name = []
tag_text = []
for tag in header_tags:
    tag_name.append(tag.name)
    tag_text.append(tag.text)

df = pd.DataFrame(data={"Tag Name":tag_name,"Tag Text":tag_text})
df
```

```python
df.to_csv("result.csv")
```
