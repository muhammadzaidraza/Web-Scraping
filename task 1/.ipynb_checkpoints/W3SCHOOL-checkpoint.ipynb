Extract all the lnks from W#school website and make excel sheet of its linka nad its text

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import xlsxwriter

# Create a workbook and add a worksheet.
workbook = xlsxwriter.Workbook('output.xlsx')
worksheet = workbook.add_worksheet()

#setting rows and column to 0
row = 0
col = 0

url = "https://www.w3schools.com/"
html = requests.get(url)
soup = BeautifulSoup(html.content, 'lxml')

print(soup.prettify())
```

```python
links = soup.find_all('a', class_="w3-bar-item w3-button")
```

```python
for lnk in links:
    text = lnk.text
    address = lnk["href"]
    worksheet.write(row, col,     text)
    worksheet.write(row, col + 1, address)
    row += 1
workbook.close()
```
